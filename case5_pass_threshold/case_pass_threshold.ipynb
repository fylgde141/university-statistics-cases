#Кейс "Проходной порог"








##Подключение библиотек

```python
import scipy.stats as sps
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

##Синий уровень

```python
p_strong = 0.98 #шанс правильно ответить на вопрос СИЛЬНОМУ кандитату
p_weak = 0.9 #шанс правильно ответить на вопрос СЛАБОМУ кандитату
n = 20 #Размер теста

#определим вероятности пройти в зависимости от проходного порога
weak = []
strong = []
for k in range(1,n+1):
  weak.append(1-sps.binom.cdf(k-1, n, p_weak))
  strong.append(1-sps.binom.cdf(k-1, n, p_strong))

df = pd.DataFrame({
    'k': list(range(1, n + 1)),
    'strong_probability': strong,
    'weak_probability': weak,
})

#строим график
plt.figure(figsize = ([10,4]))
plt.plot(df['k'], df['strong_probability'], marker = 'o',alpha = 0.7, label = 'Сильный кандидат')
plt.plot(df['k'], df['weak_probability'], marker = 'o', color = 'green', alpha = 0.7, label = 'Слабый кандидат')


plt.axhline(y=0.8, color='r', linestyle='--',alpha=0.5, label='Минимальная вероятность \nпрохождения сильного кандидата')
plt.axvline(x = 19, color = 'b', linestyle = '--', alpha = 0.7, label = 'Оптимальный проходной порог')

plt.xticks(df['k'])
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xlabel('Порог')
plt.ylabel('Вероятность пройти')
plt.title('Зависимость вероятности прохождения кандидата\n от порогового значения количества решенных задач')
plt.grid()
plt.show()
```

##Красный уровень

```python
#определим вероятности пройти в зависимости от проходного порога
data = []
for n in range(1,101):
  k =int(sps.binom.ppf(0.2, n, p_strong)) #минимально разрешимая вероятность прохождения сильного кандидата и максимальный отсев слабых кандидатов
  strong_prob = (1 - sps.binom.cdf(k-1, n, p_strong))
  weak_prob = (1 - sps.binom.cdf(k-1, n, p_weak))
  data.append({'n':n, 'strong_probability': strong_prob, 'weak_probability':weak_prob})

df_r = pd.DataFrame(data)


#строим график
plt.figure(figsize = ([10,4]))

plt.plot(df_r['n'], df_r['strong_probability'], marker = 'o',markersize = 4, label = 'Сильный кандидат')
plt.plot(df_r['n'], df_r['weak_probability'], marker = 'o',markersize = 4,alpha=0.7,  color = 'limegreen', label = 'Слабый кандидат')

plt.axhline(y=0.8, color='r', linestyle='--',alpha=0.5, label='Минимальная вероятность \nпрохождения сильного кандидата')
plt.axhline(y=0.3, color='b', linestyle='--',alpha=0.7, label='Максимальная вероятность \nпрохождения слабого кандидата')
plt.axvline(x=24, color='g', linestyle='--', label='Минимальное количество вопросов\nудволетворяющее всем требованиям')

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.xlabel('Количество задач')
plt.ylabel('Вероятность пройти')
plt.title('Зависимость вероятности прохождения кандидата\n от количества задач в тесте')
plt.xlim([0, 101])
plt.ylim([0, 1])
plt.grid()

plt.text(24, -0.066, '24', color='black', fontsize=10, ha='center')

plt.show()
```
